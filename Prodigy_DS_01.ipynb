Bar Chart and Histogram of World Population

```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style #Importing style from matplotlib
import seaborn as sns
```

```python
# Reading the CSV file to load the world population data into a pandas Dataframe
df=pd. read_csv("worldpopulation.csv")

# Displaying the DataFrame
df
```

```python
# Specifying the list of GCC countries
GCC_country = ['Kuwait', 'Bahrain', 'Oman', 'Qatar', 'Saudi Arabia', 'United Arab Emirates']
# Filtering data for the GCC countries
filtered_data = df[df['Country Name'].isin(GCC_country)]

# Specifying colors for the plot
colors = ['red', 'blue', 'green', 'orange', 'yellow', 'purple']

# Using ggplot style from matplotlib
style.use('ggplot')

# Creating the bar chart
plt.figure(figsize=(10, 5)) # Creating the figure with specified size
plt.bar(filtered_data['Country Name'], filtered_data['2020'], color=colors[:len(filtered_data)]) # Creating a bar plot with filtered data
plt.xlabel('Country Name', fontsize=12, weight='bold') # Labeling the x-axis
plt.ylabel('Population in 2020', fontsize=12, weight='bold') # Labeling the y-axis
plt.title('Population of Selected GCC Countries in 2020', fontsize=15, weight='bold') # Setting the bar chart title
plt.xticks(rotation=30, ha='right', fontsize=10) #Rotating x-axis labels for better readability
plt.yticks(fontsize=10) # Setting font size for y-axis ticks
plt.grid(axis='y', linestyle='--', alpha=0.7) # Adding grid lines for the y-axis

#Setting the background color of the plot area
plt.gca().set_facecolor('grey') 
# Display the chart

plt.show()
```

```python
# Reading the CSV file to load the world population data into a pandas Dataframe
df=pd.read_csv('worldpopulation.csv')

# Displaying the DataFrame
df
```

```python
#Filtering dataframe for Nepal
df_nepal = df[df['Country Name'] == 'Nepal'] 

df_nepal
```

```python
# Select only the columns related to years (i.e., 1960 to 2022)
population_data = df_nepal.loc[:, '1960':'2022'].values.flatten()  # Flatten to convert to a 1D array

# Ensure the data is numeric and remove any NaN values
population_data = pd.to_numeric(population_data, errors='coerce')
population_data = population_data[~pd.isna(population_data)]

# Plot the histogram
plt.figure(figsize=(10, 5))
plt.hist(population_data, bins=20, color= "blue", edgecolor='black')

# Add title and labels
plt.title('Histogram of Population Data of Nepal between (1960-2022)')
plt.yscale('log')
plt.xlabel('Population')
plt.ylabel('Frequency')

#Setting the background color of the plot area
plt.gca().set_facecolor('grey') 

# Show the plot
plt.show()
```
